```python
import random

class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def get_coord(self):
        return (self.x, self.y)
    
    def set_coord(self, x, y):
        self.x = x
        self.y = y
    
    def get_x(self):
        return self.x
    
    def set_x(self, x):
        self.x = x
        
    def get_y(self):
        return self.y
    
    def set_y(self, y):
        self.y = y
        
    def __str__(self):
        msg = f"({self.x}, {self.y})"
        return msg
        
    def __eq__(self, other):
        if self.x == other.x and self.y == other.y:
            return True
        else: 
            return False
```

```python
class Player:
    def __init__(self, name, x=0, y=0):
        # 이름과 좌표를 멤버변수로 저장합니다. 
        self.name = name
        self.point = Point2D(x, y)

    def check_boundary_out(self):
        # 조건문을 활용해서 현재 플레이어의 좌표가 (0, 0), (5, 5) 내에 있는지 검사해서
        # 밖에 있다면 True를 반환하고 위의 그림을 참고하여 메세지를 출력합니다. 
        # boundary_in이라면 False를 반환합니다.
        if self.point.x > 5 or self.point.y > 5 or self.point.x < 0 or self.point.y <0:
            print("갈 수 없는 곳입니다.")
            return True
        else:
            return False

    def move_right(self):
        # Point2D 클래스의 set 함수를 활용하여 좌표를 변경합니다. 
        # 좌표를 이동시킨 후 check_boundary_out으로 좌표가 격자 내에 있는지 검사합니다.
        # 검사를 했는데, False라면 다시 원래의 좌표로 되돌립니다. 
        # 아래의 left, up, down 함수도 마찬가지로 구현합니다. 
        self.point.set_x(self.point.x + 1)
        if self.check_boundary_out():
            self.point.set_x(self.point.x - 1)

    def move_left(self):
        self.point.set_x(self.point.x - 1)
        if self.check_boundary_out():
            self.point.set_x(self.point.x + 1)

    def move_up(self):
        self.point.set_y(self.point.y + 1)
        if self.check_boundary_out():
            self.point.set_y(self.point.y - 1)
            
    def move_down(self):
        self.point.set_y(self.point.y - 1)
        if self.check_boundary_out():
            self.point.set_y(self.point.y + 1)

    def __str__(self):
        # 아래 문장을 참고해서 메세지를 만들고 반환하세요
        # 플레이어 : 김국진, 현재위치 : [0, 5]
        msg = f"플레이어 : {self.name}, 현재위치 {self.point.get_coord()}"
        return msg
```

```python
class Treasure:
    def __init__(self, name):
        # name을 멤버변수로 저장합니다.
        # random.randint()를 활용해서 보물의 Point를 할당합니다. 
        # 구현 : 
        self.name = name
        x = random.randint(1,5)
        y = random.randint(1,5)
        self.point = Point2D(x, y)

    def get_name(self):
        return self.name
    
    def __str__(self):
        # 아래 문장을 참고해서 메세지를 만들고 반환하세요
        # 보물 : 금, 보물위치 : (3,3)
        msg = f"보물 : {self.name}, 보물위치 : {self.point.get_coord()}"
        return msg
```

```python
def check_player_position(player, gold):
    # 멤버접근 연산자와 이미 구현된 __eq__ 연산자 오버로딩을 활용하여
    # 조건문을 만들고 두 좌표가 같으면 True를 반환하고 위와 같은 형식으로 메세지를 출력하세요
    # 구현 : 
    if player.point == gold.point:
        # "Player"이 "금"을 찾았습니다!
        print()
        print(f'플레이어" {player.name}이(가) "{gold.name}"를 찾았습니다!') # 구현
        return True
    else:
        return False
```

```python
def run_game():
    found = False
    name = input("플레이어 이름을 입력하세요. : ")# 구현 : input() 함수를 이용해 위 그림과 같이 이름을 입력받으세요
    player = Player(name, x=0, y=0)# 구현 : Player 클래스를 활용해서 player 인스턴스를 생성하세요
    gold = Treasure('머신러닝 비법서')# 구현 :  Treasure 클래스를 활용해서 gold 인스턴스를 생성하세요
    print(gold)
    
    while not found:
        direction = input("이동할 방향을 입력하세요 : ")# input() 함수를 이용해 위 그림과 같이 WASD 중... 문장으로 입력을 받으세요
        # 조건문을 활용해서 입력받은 character에 따라 플레이어의 움직임을 구현하세요 
        if direction.lower() == "w":
            player.move_up()
            print(player)

        elif direction.lower() == "s":
            player.move_down()
            print(player)

        elif direction.lower() == "a":
            player.move_left()
            print(player)

        elif direction.lower() == "d":
            player.move_right()
            print(player)

        else:
            print("잘못된 키 입력입니다. WASD 중 하나를 입력하세요")
        
        if check_player_position(player, gold):
            found = True
        print()
```

```python
run_game()
```
